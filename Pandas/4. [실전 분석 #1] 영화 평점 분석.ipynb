## [실습 #4] 연령대 별로 영화 평점 분석하기
연령대(10대 미만, 10대, 20대, ...50대) 컬럼을 추가한 후, 영화별 연령대별 영화평점 구하기

```python
data
```

```python
# 함수 정의
def generate_ages(X):
    if X < 10:
            return '10대 미만'
    elif X < 20:
            return '10대'
    elif X < 30:
            return '20대'
    elif X < 40:    
            return '30대'
    elif X < 50:  
            return '40대'
    else:
            return '50대 이상'
```

```python
data['연령대'] = data.연령.apply(generate_ages)
```

```python
data
```

```python
ex4 = data.pivot_table(index = '영화제목', columns = '연령대',
                aggfunc = 'mean', values = '평점')
```
