# 1. 맥주 데이터를 활용한 단순 선형 회귀 실습

### 데이터 수집

```python
beer = pd.read_csv("./data/beer.csv")
beer
```

```python
beer.pivot_table(index = 'temperature', values = 'beer', aggfunc = 'mean').plot()
```

```python
beer
```

### 데이터 탐색

```python
plt.figure(figsize = (5,5))
plt.scatter(beer['temperature'], beer['beer'])
plt.xlabel('temperature')
plt.ylabel('beer')
plt.grid()
plt.show()
```

```python
# 독립 변수, 종속 변수 확인
beer['temperature']
```

```python
# 종속 변수 확인, 레코드 수 확인
```

```python
print(beer['beer'])
print(len(beer))
```

###  데이터 준비 - 학습용 / 검증용 데이터 분리

```python
# 전체 데이터 중 80%는 학습용, 20%는 검증용으로 분리
import numpy as np

#독립변수, 종속변수 데이터셋 준비
X = np.array(beer['temperature']).reshape(-1, 1)
y = beer['beer']

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
len(X_test)
```

```python
y_test
```

### 모델 구축

```python
#모델 클래스 선택 후 인스턴스 객체 생성
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
```

```python
# 학습 수행
reg = lr.fit(X_train,y_train)
```

```python
# 계수 및 절편 확인: _속성은 학습을 통해 결정되는 속성
reg.coef_, reg.intercept_
```

```python
#회귀식
print("y = {:.2f}X + {:.3f}".format(reg.coef_[0], reg.intercept_))
```

```python
# 구축된 모델에서 예측 수행
y_pred = reg.predict(X_test)
print(np.round(y_pred, 2))
```

### 모델 성능 평가

```python
from sklearn.metrics import mean_squared_error, r2_score
#MSE
mse = mean_squared_error(y_test, y_pred)

#RMSE
rmse = np.sqrt(mse)

print("MSE:", np.round(mse,3))
print("RMSE: ", np.round(rmse, 3))
```

```python
#결정계수 R2
r2 = r2_score(y_test, y_pred)
print("R2: ", np.round(r2, 3))
```

### 시각화

```python
#y = 0.697056X + 36.06 시각화

plt.figure(figsize=(5,5))
xx = np.arange(beer['temperature'].min() - 1,
               beer['temperature'].max() + 1)
yy = reg.predict(xx.reshape(len(xx), 1))

plt.plot(xx, yy, linestyle='--', color='red')
# 수집한 데이터셋 시각화
plt.scatter(beer['temperature'], beer['beer'])
plt.xlabel('temperature')
plt.ylabel('beer')
plt.grid()
plt.show()
```

# 응용1. 나이, 키의 관계를 회귀 분석

```python
# 필요한 모듈 import 
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

#### 데이터 생성/수집

```python
data_df = pd.read_csv("./data/age_height.csv" )
```

```python
data_df
```

```python
data_df['age(X)'].value_counts()
```

```python
data_df.set_index('age(X)').plot(kind = 'bar')
```

#### 데이터 탐색

```python
# 데이터 분포 시각화
```

```python
plt.figure(figsize = (5,5))
plt.scatter(data_df['age(X)'], data_df['height(T)'])
plt.xlabel('age(X)')
plt.ylabel('height(T)')
plt.grid()
plt.show()
```

#### 데이터 준비 - 학습용 / 검증용 데이터 분리

```python
# 전체 데이터 중 80%는 학습용, 20%는 검증용으로 분리

X = np.array(data_df['age(X)']).reshape(-1, 1)
y = data_df['height(T)']

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)
```

```python
#####모델 구축 및 예측
```

```python
#모델 객체 생성  및 학습
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
reg = lr.fit(X_train,y_train)
# 예측 수행
y_pred = reg.predict(X_test)
print(np.round(y_pred, 2))
```

```python
#####모델 성능 평가

#MSE : metrcics 모듈의 mean_squared_error() 함수 사용
#RMSE : numpy의 sqrt() 함수 사용

#MSE
mse = mean_squared_error(y_test, y_pred)

#RMSE
rmse = np.sqrt(mse)

print("MSE:", np.round(mse,3))
print("RMSE: ", np.round(rmse, 3))


#결정계수 R2
r2 = r2_score(y_test, y_pred)
print("R2: ", np.round(r2, 3))
print("coef: ", reg.coef_)
print("intercept: ", reg.intercept_)
#회귀식
print("y = {:.2f}X + {:.3f}".format(reg.coef_[0], reg.intercept_))
```

```python
#나이, 키 데이터 실제 데이터 및 회귀식 시각화

#y = 1.44X + 138.784시각화
plt.figure(figsize=(5,5))
xx = np.arange(data_df['age(X)'].min() - 1,
               data_df['age(X)'].max() + 1)
yy = reg.predict(xx.reshape(len(xx), 1))

plt.plot(xx, yy, linestyle='--', color='red')
# 수집한 데이터셋 시각화

plt.scatter(data_df['age(X)'], data_df['height(T)'])
plt.xlabel('age(X)')
plt.ylabel('height(T)')
plt.grid()
plt.show()
```


# < 스탯츠모델에서 단순선형회귀 수행>

```python
#스탯츠모델 import
import statsmodels.api as sm

# 맥주 데이터 
# 데이터 생성
beer = pd.read_csv("./data/beer.csv")

#독립변수, 종속변수 데이터셋 준비
import numpy as np

X = np.array(beer['temperature']).reshape(-1, 1)
y = beer['beer']

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 1)

```

```python
X_test
```

```python
sm.add_constant(X_test)
```

```python
# 스탯츠모델
X_train = sm.add_constant(X_train)
X_test = sm.add_constant(X_test)

lr = sm.OLS(y_train, X_train)

reg = lr.fit()
```

```python
# 성능 평가 Prob 값이 0.05보다 무조건 작아야함, 각 계수의 P값이 0.05보다 크면 안됨
print(reg.summary())
```

```python
import pandas as pd

df = pd.DataFrame(boston.data, columns = boston.feature_names)
df 
```

```python
X = pd.DataFrame(df['RM'])
y = boston.target
```

```python
X
```

```python
plt.figure(figsize = (5,5))
plt.scatter(df['RM'], y)
plt.xlabel('방의개수')
plt.ylabel('가격')
plt.grid()
plt.show()
```

### 학습용/ 검증용 데이터 분리

```python
X
```

```python
y
```

```python
# 조건 : 학습 7, 검증 3
# seed  = 1로 고정
##### 5분 실습 / 결정 계수를 구하라

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 1)
```

### 모델 구축

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression().fit(X_train,y_train)
```

### 모델 평가

```python
y_pred = reg.predict(X_test)

#MSE
mse = mean_squared_error(y_test, y_pred)

#RMSE
rmse = np.sqrt(mse)

print("MSE:", np.round(mse,3))
print("RMSE: ", np.round(rmse, 3))


#결정계수 R2
r2 = r2_score(y_test, y_pred)
print("R2: ", np.round(r2, 3))
print("coef: ", reg.coef_)
print("intercept: ", reg.intercept_)
#회귀식
print("y = {:.2f}X + {:.3f}".format(reg.coef_[0], reg.intercept_))
```

### 시각화

```python
plt.figure(figsize=(5,5))
xx = np.linspace(X.min(), X.max())
yy = reg.predict(xx.reshape(len(xx), 1))

plt.plot(xx, yy, linestyle='--', color='green')
# 수집한 데이터셋 시각화


plt.scatter(df['RM'], y, c='r', s=5)
plt.xlabel('# of rooms')
plt.ylabel('price')
plt.grid()
plt.show()
```
